# Linear regression 
i this task i will show to fit a model to a data using linear regression

```python
#import libraries 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import math
```

```python
#TODO read point from txt
file = open('case1.txt')
```

```python
#parsing the file
x = []
y = []
for i in file:
    point = i.rstrip().split(',')
    x.append(float(point[0]))
    y.append(float(point[1]))
```

visualize the data to get an intuition how the data looks like

```python
#plot the data
plt.scatter(x,y, s=1)
plt.show()
```

## Removing outliers
i removed the outliers manually by calculating the mean and the standard deviation
then any point more than 2 standard deviation from the mean and less than 2 standard deviation

```python
#remove the outliers
mean = np.mean(y) #mean
std = np.std(y) #standard deviation
#remove the point from the x and y
i = 0 #counter 
while i != len(y):
    if (y[i] <= mean - 2*std) or (y[i] >= mean + 2*std):
        y.pop(i)
        x.pop(i)
    else:
        i = i+1
```

Here is the data after removing outliers

```python
#plot the data
plt.scatter(x,y, s=1)
plt.show()
```

## model 
i decided to use the sckitlearn library to make linear regression 


```python
#change the x,y into numpy array, because the fit method take the estimator and response as numpy arraies
x = np.array(x).reshape((-1, 1))
y = np.array(y)
```

```python
#creating a model
model = LinearRegression()
model.fit(x, y)
```

```python
#calculate coefficient of determination R^2 of the prediction.
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```

## Plot the model

```python
#draw the line
y_hat = model.predict(x)
plt.scatter(x,y, s=1)
plt.plot(x,y_hat,color=(0,1,0))
```

## Confidence interval
For 99.9% confidence level z= 3.291

```python
#Calculate the confidence interval of error for 99.9% confidence level.
mean = np.mean(y) #mean
std = np.std(y) #standard deviation
st_error = std / math.sqrt(y.size) #standard error
# For 99.9% confidence level z= 3.291
z = 3.291
ME = z * st_error #Margin error
print("Mean value: ",mean)
print("confidence interval: ", mean+ME, " , ", mean-ME)
```
